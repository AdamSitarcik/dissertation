```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
A=186

def e1(e):
    return 6.76*A**(-2/3)*e**(-3)*10**(15)
def m1(e):
    return 2.2*e**(-3)*10**(15)
def e2(e):
    return 9.52*A**(-4/3)*e**(-5)*10**(15)
def m2(e):
    return 3.1*A**(-2/3)*e**(-5)*10**(15)
def e3(e):
    return 2.04*A**(-2)*e**(-7)*10**(15)
def m3(e):
    return 6.66*A**(-4/3)*e**(-7)*10**(15)
def e4(e):
    return 6.5*A**(-8/3)*e**(-9)*10**(15)
def m4(e):
    return 2.12*A**(-2)*e**(-9)*10**(15)

x = np.linspace(10, 1000, 1000)

fig, ax = plt.subplots(figsize=(7,4))

ax.plot(x, e1(x), color='black')
ax.plot(x, m1(x), color='black', linestyle='dashed')
ax.plot(x, e2(x), color='red')
ax.plot(x, m2(x), color='red', linestyle='dashed')
ax.plot(x, e3(x), color='blue')
ax.plot(x, m3(x), color='blue', linestyle='dashed')
ax.plot(x, e4(x), color='green')
ax.plot(x, m4(x), color='green', linestyle='dashed')

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(10,1000)

ax.set_ylabel('Half-life [s]', size=15)
ax.set_xlabel('Energy of a transition [keV]', size=15)

plt.xticks(size=12)
plt.yticks(size=12)

ax.xaxis.set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax.tick_params(axis='both', which='both', direction='in')
ax.tick_params(axis='both', which='major', size=6)
ax.tick_params(axis='both', which='minor', size=3)

plt.savefig('weisskopf_halflives.svg', bbox_inches='tight')
plt.show()
```
